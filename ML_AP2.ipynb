```python
# Corrigir tipos de dados das colunas numéricas que estão como objeto (texto)
numeric_columns = ["Respiratory Rate (breaths/min)", "Heart Rate (rates/min"]
for col in numeric_columns:
    train_data_cleaned[col] = pd.to_numeric(train_data_cleaned[col], errors="coerce")
```

```python
# Remover colunas irrelevantes (nomes de pacientes e pais, ID, nome do instituto)
columns_to_drop = [
    "Patient Id", "Patient First Name", "Family Name", "Father's name",
    "Institute Name", "Location of Institute"
]
train_data_cleaned = train_data.drop(columns=columns_to_drop)
```
