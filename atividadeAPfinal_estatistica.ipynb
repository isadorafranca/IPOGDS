```python
# 1. Importar bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 3. Ler o arquivo CSV (substitua o nome se necess√°rio)
df = pd.read_csv("AP5.csv")
```

```python
# 4. Visualizar os dados
print(df.head())
print("\nInforma√ß√µes gerais:")
print(df.info())
```

```python
# 5. An√°lise estat√≠stica b√°sica
print("\nEstat√≠sticas descritivas:")
print(df.describe())
```

```python
# Selecionar uma coluna num√©rica
coluna = 'Hours_of_VR_Usage_Per_Week'  # Exemplo de coluna num√©rica

media = df[coluna].mean()
mediana = df[coluna].median()
moda = df[coluna].mode()[0]
desvio_padrao = df[coluna].std()

print(f"\nM√©dia: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Desvio-padr√£o: {desvio_padrao:.2f}")
```

```python
# 6. Gr√°ficos
sns.set(style="whitegrid")

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(y=df[coluna])
plt.title(f'Boxplot - {coluna}')
plt.show()
```

```python
# Histograma
plt.figure(figsize=(6,4))
sns.histplot(df[coluna], bins=10, kde=True)
plt.title(f'Histograma - {coluna}')
plt.xlabel(coluna)
plt.show()
```

```python
# Gr√°fico de linhas (exemplo com √≠ndice)
plt.figure(figsize=(6,4))
sns.lineplot(data=df[coluna])
plt.title(f'Gr√°fico de Linhas - {coluna}')
plt.ylabel(coluna)
plt.xlabel('√çndice')
plt.show()
```
