```python
# 1. Importar bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 2. Carregar dados (após fazer upload do arquivo no Colab)
from google.colab import files
uploaded = files.upload()
```

```python
# 3. Ler o arquivo CSV (substitua o nome se necessário)
df = pd.read_csv("AP5.csv")
```

```python
# 4. Visualizar os dados
print(df.head())
print("\nInformações gerais:")
print(df.info())
```

```python
# 5. Análise estatística básica
print("\nEstatísticas descritivas:")
print(df.describe())
```

```python
# Selecionar uma coluna numérica
coluna = 'Hours_of_VR_Usage_Per_Week'  # Exemplo de coluna numérica

media = df[coluna].mean()
mediana = df[coluna].median()
moda = df[coluna].mode()[0]
desvio_padrao = df[coluna].std()

print(f"\nMédia: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Desvio-padrão: {desvio_padrao:.2f}")
```

```python
# 6. Gráficos
sns.set(style="whitegrid")

# Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(y=df[coluna])
plt.title(f'Boxplot - {coluna}')
plt.show()
```

```python
# Histograma
plt.figure(figsize=(6,4))
sns.histplot(df[coluna], bins=10, kde=True)
plt.title(f'Histograma - {coluna}')
plt.xlabel(coluna)
plt.show()
```

```python
# Gráfico de linhas (exemplo com índice)
plt.figure(figsize=(6,4))
sns.lineplot(data=df[coluna])
plt.title(f'Gráfico de Linhas - {coluna}')
plt.ylabel(coluna)
plt.xlabel('Índice')
plt.show()
```
