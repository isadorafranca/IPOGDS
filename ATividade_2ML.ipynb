# **Limpeza de dados:**

*   Verificar e tratar valores ausentes
*   Verificar e tratar duplicatas



```python
# Verificar e tratar valores ausentes
print("Valores ausentes por coluna antes do tratamento:")
print(df.isnull().sum())
```

```python
#Preencher valores ausentes com medianas (colunas num√©ricas)
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
##Preencher valores ausentes com moda(colunas num√©ricas)
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#duplicatas
df.drop_duplicates(inplace=True)
```

```python
#outliers para colunas num√©ricas
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
     # Substituir outliers pelos limites
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
```

```python
print("\nValores ausentes por coluna ap√≥s o tratamento:")
print(df.isnull().sum())
```

# **Agora vamos converter os dados**


*   convers√£o de colunas categ√≥ricas bin¬¥¬¥arias
*   Convers√£o outras colunas categ√≥ricas usando Label Encoding
*   Convers√£o colunas de sintomas para valores num√©ricos consistentes






```python
# Converter colunas categ√≥ricas bin√°rias para num√©ricas (0/1)
#para saber a lista de colunas print(df.columns)
binary_cols = ['Genes in mother\'s side', 'Inherited from father', 'Maternal gene',
               'Paternal gene', 'Parental consent', 'Gender', 'Birth asphyxia',
               'Autopsy shows birth defect (if applicable)', 'Folic acid details (peri-conceptional)',
               'H/O serious maternal illness', 'H/O radiation exposure (x-ray)',
               'H/O substance abuse', 'Assisted conception IVF/ART',
               'History of anomalies in previous pregnancies']

for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].map({'Yes': 1, 'No': 0, 'Not applicable': np.nan})
        # checando valores vazios
        mode_values = df[col].mode()
        if not mode_values.empty:
            df[col].fillna(mode_values[0], inplace=True)

        else:
            df[col].fillna(0, inplace=True)
```

```python
# Converter outras colunas categ√≥ricas usando Label Encoding
label_encoder = LabelEncoder()
for col in categorical_cols:
    if df[col].nunique() < 50 and col not in binary_cols:  # Aplicar apenas para colunas com poucas categorias
        df[col] = label_encoder.fit_transform(df[col].astype(str))
```

```python
# Converter colunas de sintomas para valores num√©ricos consistentes
symptom_cols = ['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5']
for col in symptom_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(0, inplace=True)  # Assumir que valores ausentes significam sintoma ausente
```

# **Tranformando dados**


*   padroniza√ß√£o de colunas
*   codifica√ß√£o one-hot



```python
# Normaliza√ß√£o/padroniza√ß√£o de colunas num√©ricas
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
#Codifica√ß√£o one-hot para algumas colunas categ√≥ricas selecionadas
# (Aplicar apenas para colunas categ√≥ricas com poucas categorias)
one_hot_cols = ['Status', 'Respiratory Rate (breaths/min)', 'Heart Rate (rates/min',
                'Blood test result', 'Genetic Disorder', 'Disorder Subclass']
for col in one_hot_cols:
    if col in df.columns:
        dummies = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, dummies], axis=1)
        df.drop(col, axis=1, inplace=True)
```

```python
# Engenharia de caracter√≠sticas
if 'Patient Age' in df.columns:
    df['Age_Group'] = pd.cut(df['Patient Age'],
                            bins=[0, 2, 12, 18, 60, 100],
                            labels=['Infant', 'Child', 'Teen', 'Adult', 'Elderly'])

# Salvar o dataset limpo e transformado
df.to_csv('cleaned_train.csv', index=False)

print("\nLimpeza, convers√£o e transforma√ß√£o conclu√≠das com sucesso!")
print(f"Dataset final tem {df.shape[0]} linhas e {df.shape[1]} colunas.")
```
